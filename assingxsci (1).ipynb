# Extracting the only state name from 'state' coloumn by removing 'state_name:' prefix .

```python
def extract_state_name(df):
  state = df['state']
  state_name = state.split('state_name:')[1]

  return state_name

df['State'] = df.apply(extract_state_name, axis=1)

df
```

# Extracting the only district name from 'District_name' coloumn by removing 'district_name:' prefix .

```python
def extract_district_name(df):
  district = df['District_name']
  district_name = district.split('district_name:')[1]

  return district_name

df['district_name']=df.apply(extract_district_name, axis=1)

df
```

```python
# Drop 'District_name' column
df=df.drop('District_name',axis=1)
```

```python
# frop 'state' column
df=df.drop('state',axis=1)
df
```

# Extracting the only district code  from 'District_code' coloumn by removing 'district_code:' prefix .

```python
def extract_district_code(df):
  district_c = df['District_code']
  district_code = district_c.split('district_code:')[1]

  return district_code

df['district_code '] = df.apply(extract_district_code, axis=1)

df
```

# Sub Assignment 2: Exploring Indian States Data

```python
# importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```





# 1 State wise population of each *state*

```python
# Get the state wise population
state_wise_population = df.groupby('State')[['Population']].sum()
```

```python
state_wise_population
```

# State wise population in bar chart

```python
plt.bar(df["State"], df["Population"])
plt.xlabel("State")
plt.ylabel("Population")
plt.title("Population of Indian States")
plt.show()
```

```python
# Group the DataFrame by state and calculate the total population
grouped_df = df.groupby('State')['Population'].sum()

# Get pie chart slice labels
pie_chart_slice_labels = grouped_df.index.to_numpy()

# Remove slice labels beyond the top 15 to avoid labels overlapping
pie_chart_slice_labels[15:] = ''

# Get pie chart slice values
pie_chart_slice_values = grouped_df.to_numpy()

# Create a pie chart of the state wise population
plt.pie(pie_chart_slice_values, labels=pie_chart_slice_labels, autopct="%1.1f%%")
plt.title("State Wise Population in India")
plt.show()
```

#2  What is the average literacy rate in India?

```python
# Calculate the average literacy rate
average_literacy_rate = (df["Literate"] / df["Population"] * 100).mean()

# Print the average literacy rate
average_literacy_rate
```

# State wise average litracy rate visualization

```python
# Create a bar graph of the average literacy rate
plt.bar(df['State'], average_literacy_rate)
plt.xlabel("State")
plt.ylabel("Average Literacy Rate")
plt.title("Average Literacy Rate in India")
plt.show()
```

# 3 Average literacy rate for males

```python

# Calculate the average literacy rate for females
average_female_literacy_rate = (df["Female_Literate"] / df["Female"] * 100).mean()
average_female_literacy_rate
```
